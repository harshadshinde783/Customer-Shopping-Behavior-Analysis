```python
# fill null value in review rating by category group 

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.mean))
```

```python
df.isnull().sum()
```

```python
# data are loading into mysql to better understad business probelm 
# convert columns into small name 
# and inbetween use 'underscore' 


df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)'  : 'purchase_amount'})
```

```python
df.columns
```

```python
# create a new columns name as age_group 


labels = ['Young-Adult','Adult','Middle-aged','Senior']

df['age_group'] = pd.qcut(df['age'],q=4, labels=labels)

```

```python
df[['age_group']].value_counts()
```

```python
df[['frequency_of_purchases']].value_counts()
```

```python
# create columns purchase_frequency_days


frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
